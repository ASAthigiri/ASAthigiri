Ex: 1 IMPLEMENT SIMPLE PROGRAMS LIKE VECTOR ADDITION IN TENSORFLOW


```python
import tensorflow as tf
vector1 = tf.constant([2.0, 3.0, 4.0])
print(vector1)
```

```python
vector2 = tf.constant([3.0, 4.0, 5.0])
print(vector2)
```

```python
print(tf.add(vector1,vector2))
```

```python
print(vector1*vector2)
```

```python
print(tf.tensordot(vector1,vector2,axes=1))
```

```python
a = tf.constant([[1., 2., 3.], [4., 5., 6.], [7., 8., 9.]])
b = tf.constant([1., 2., 3.])
c = tf.tensordot(a, b, axes=1)
print(c)
```
